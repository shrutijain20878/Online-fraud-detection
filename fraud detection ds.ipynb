```python
import pandas as pd
import numpy as np
df=pd.read_csv("new1.csv")
print(df)
```

```python
del df['isFlaggedFraud']
```

```python
df.isnull().sum()
```

```python
boo=pd.isnull(df["amount"])
df[boo]
```

```python
df.fillna(0)
```

```python
df.fillna(method="pad")## take the adjacent upper value 
```

```python
df.fillna(method="ffill")# fill forward column
```

```python
df.fillna(method="bfill")# fill backward column
```

```python
df["nameDest"].fillna(" ",inplace=True)
df
```

```python
df["nameOrig"].fillna("NaN",inplace=True)
df
```

```python
df=df.fillna(method="ffill")
df.isnull().sum()
df
```

```python
df.replace(to_replace=np.nan,value=-99)##only chnge in numerical data not in string data
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
labels=label_encoder.fit_transform(df.type)
df['type']=labels
print(df)
df=df.drop(['nameOrig','nameDest'],axis=1)
print(df)
X=df.iloc[:,:-1].values#independent
Y=df.iloc[:,-1].values#dependent
print(X)
print(Y)
```

```python
#default 75 25 train test
# split of tain test data 70-30 or 80-20 and 60-20-20 or 70-15-15 for train-validation-test split dataset
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
print('X_train.shape:',X_train.shape)
print('X_test.shape:',X_test.shape)
print('Y_train.shape:',Y_train.shape)
print('Y_test.shape:',Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()#constructor calling
regressor.fit(X_train,Y_train)
```

```python
y_pred=regressor.predict(X_test)
Y_test-y_pred
```

```python
print(regressor.score(X_test,Y_test))##calculate the accuracy
```

```python
from sklearn.metrics import mean_squared_error,r2_score
rmse=(np.sqrt(mean_squared_error(Y_test,y_pred)))
r2score=r2_score(Y_test,y_pred)
print(rmse)
print(r2score)
```
